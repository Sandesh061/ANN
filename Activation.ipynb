```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,20)
plt.plot(x,x)
plt.title("Linear activation function")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x=np.linspace(-10,20)
plt.plot(x, 1/(1+np.exp(-x)))
plt.title("Sigmoid Activation function")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x=np.linspace(-10,20)
plt.plot(x, np.tanh(x))
plt.title("Tanh Activation function")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x=np.linspace(-10,20)
plt.plot(x, np.exp(x)/np.sum(np.exp(x)))
plt.title("Softmax Activation function")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x1=[]
for i in x:
    if i <0:
        x1.append(0)
    else:
        x1.append(i)
x=np.linspace(-10,20)
plt.plot(x,x1)
plt.title("Relu Activation function")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
